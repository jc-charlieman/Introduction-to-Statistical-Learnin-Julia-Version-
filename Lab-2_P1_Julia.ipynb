```julia
rng = MersenneTwister(1)
y = rand!(rng, rand(Normal(), 10))
mean(y)
```

```julia
y
```

```julia
# Sample variance
var(y), sum((y .- mean(y)).^2)/(length(y)-1)
```

```julia
# Population variance
var(y, corrected=false), sum((y .- mean(y)).^2)/(length(y))
```

```julia
sqrt(var(y)), std(y)
```

```julia
scatter(x, y)
```

```julia
scatter(x, y, xlabel="this is the x-axis", ylabel="this is the y-axis", 
title="Plot of X vs Y", size=(800, 600))
```

```julia
plot(layout=(2, 3), size=(1300, 350))
```

```julia
p1 = scatter(); p3 = scatter(); p4 = scatter(); p5 = scatter()
p2 = scatter(x, y, markershape=:circle)
p6 = scatter(x, y, markershape=:+)
plot(p1, p2, p3, p4, p5, p6, layout=(2, 3), size=(1300, 350))
```

#### Boolean indexing

```julia
keep_rows = Bool.(zeros(size(A)[1]))
```

```julia
keep_rows[[2, 4]] .= true
keep_rows
```

```julia
keep_rows == [false, true, false, true]
```

```julia
A[[1,2,1,2], :]
```

```julia
A[keep_rows, :]
```

```julia
keep_cols = Bool.(zeros(size(A)[2]))
keep_cols[[1, 3, 4]] .= true
A[keep_rows, keep_cols]
```
